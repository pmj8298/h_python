```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints) #(0,0)에서 (6,250)까지 직선을 그림
plt.show()
```

# Matplotlib Plotting

plot()함수는 다이어그램에 점(마커)을 그리는 데 사용됩니다.

기본적으로 이 plot()함수는 지점 간에 선을 그립니다.

이 함수는 다이어그램의 점을 지정하기 위한 매개변수를 사용합니다.

매개변수 1은 x축의 점을 포함하는 배열입니다 .

매개변수 2는 y축의 점을 포함하는 배열입니다 .

(1, 3)에서 (8, 10)까지 선을 그려야 하는 경우, plot 함수에 두 개의 배열 [1, 8]과 [3, 10]을 전달해야 합니다.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## Plotting Without Line

To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

## Multiple Points

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show() # Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10)
```

## Default X-Points

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show() # they will get the default values 0, 1, 2, 3 etc.
```

# Matplotlib Markers

## Markers
to emphasize each point with a specified marker

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints, marker = '*')
plt.show()
```

## Format Strings fmt

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

- 'o' → 각 데이터 포인트에 동그라미(o) 마커 추가
- ':' → 점선 스타일 적용
- 'r' → 빨간색(red)으로 선 색상 지정

## Marker Size

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

## Marker Color

### markeredgecolor or the shorter mec

```python
# markeredgecolor or the shorter mec
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'm')
plt.show()
```

### markerfacecolor or the shorter mfc

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'm')
# plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```
