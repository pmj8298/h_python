# Matplotlib Plotting

plot()함수는 다이어그램에 점(마커)을 그리는 데 사용됩니다.

기본적으로 이 plot()함수는 지점 간에 선을 그립니다.

이 함수는 다이어그램의 점을 지정하기 위한 매개변수를 사용합니다.

매개변수 1은 x축의 점을 포함하는 배열입니다 .

매개변수 2는 y축의 점을 포함하는 배열입니다 .

(1, 3)에서 (8, 10)까지 선을 그려야 하는 경우, plot 함수에 두 개의 배열 [1, 8]과 [3, 10]을 전달해야 합니다.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## Create a Title for a Plot

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

## Set Font Properties for Title and Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

## ColorMap 

- 키워드 인수에 cmap색상맵의 값을 지정하여 색상맵을 지정할 수 있습니다. 이 경우 색상 'viridis'맵은 Matplotlib에서 제공하는 기본 제공 색상맵 중 하나입니다.
또한 산점도의 각 지점에 대해 하나의 값, 즉 0~100까지의 값을 갖는 배열을 생성
- 보라색인 0에서 노란색인 100까지 범위

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar() # 컬러바 표현되는 문장

plt.show()
```

## Size 

- 인수를 사용하여 점의 크기를 변경할 수 있습니다 s.
- 색상과 마찬가지로 크기에 대한 배열의 길이가 x축과 y축에 대한 배열의 길이와 같아야 합니다

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

## alpha 

- 인수를 사용하여 점의 투명도를 조절

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

## Combine Color Size and Alpha

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

## startangle

- startangle변수는 각도로 정의되며 기본 각도는 0
